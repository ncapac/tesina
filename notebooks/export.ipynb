```python
from google.colab import drive
from pathlib import Path
import shutil

drive.mount("/content/drive")

save_root = Path("/content/drive/MyDrive/tesina_artifacts")
save_ckpt = save_root / "checkpoints"
save_exp = save_root / "exports"
save_ckpt.mkdir(parents=True, exist_ok=True)
save_exp.mkdir(parents=True, exist_ok=True)

ckpt_dir = Path("/content/tesina/checkpoints")
export_dir = Path("/content/tesina/results/exports")

candidates = [
ckpt_dir / "rf_best_model.pkl",
sorted(ckpt_dir.glob("rf_ckpt_epoch.pk")),
sorted(export_dir.glob("rf_baseline_.tar.gz")),
sorted(export_dir.glob("rf_baseline_.manifest.json")),
]

copied = []
missing = []

for src in candidates:
    if isinstance(src, list):
        for s in src:
            if s.exists():
                dst = save_ckpt / s.name if s.parent == ckpt_dir else save_exp / s.name
                shutil.copy2(s, dst)
                copied.append(dst)
                print("copied:", s, "->", dst)
            else:
                missing.append(s)
    else:
        if src.exists():
            dst = save_ckpt / src.name if src.parent == ckpt_dir else save_exp / src.name
            shutil.copy2(src, dst)
            copied.append(dst)
            print("copied:", src, "->", dst)
        else:
            missing.append(src)
print("\nSaved under:", save_root)
print("Copied files:", len(copied))
```
